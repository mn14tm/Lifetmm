# Solve Waveguide Modes for planar waveguide
Configuration is a core with two claddings above and below.

See Quantum Electronics by Amnon Yariv chapter 22 for equations

```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from numpy import pi, tan
from scipy.constants import c
%matplotlib notebook
```

```python
# input parameters
n1 = 1
n2 = 3.48
n3 = 1

lam = 1.55*1E-6
omega = 2*pi/lam
k = omega / c

t = 1*lam
```

```python
beta = np.linspace(k*n3, k*n2, num=500, endpoint=False)
```

```python
h = sp.sqrt((n2*k)**2 - beta**2)
q = sp.sqrt(beta**2 - (n1*k)**2)
p = sp.sqrt(beta**2 - (n3*k)**2)
```

```python
lhs = tan(h*t)
rhs = (q+p) / (h*(1-(p*q)/h**2))
plt.plot(lhs)
plt.plot(rhs)
```
